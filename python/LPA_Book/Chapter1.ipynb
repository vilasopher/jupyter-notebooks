```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.random.random(10000)
plt.hist(x, density=True, bins=50, cumulative=False)
plt.show()
plt.hist(x, density=True, bins=50, cumulative=True)
plt.show()

y = [x**2 for x in x]
plt.hist(y, density=True, bins=50, cumulative=False)
plt.show()
plt.hist(y, density=True, bins=50, cumulative=True)
plt.show()
```

```python
x = np.random.exponential(size=10000)
plt.hist(x, density=True, bins=50, cumulative=False)
plt.show()

y = np.array([ np.sum(x[:n]) / n for n in range(1,10000) ])
plt.hist(y, density=True, bins=50, cumulative=False)
plt.show()

plt.plot(y)
plt.show()
```

```python
def sample_sum(n):
    x = np.random.exponential(size=n)
    return np.sum(x) / n
```

```python
x = np.array([sample_sum(100) for _ in range(10000)])
plt.hist(x, density=True, bins=50, cumulative=False)
plt.show()

x = np.array([sample_sum(10000) for _ in range(10000)])
plt.hist(x, density=True, bins=50, cumulative=False)
plt.show()
```

```python
def z(n):
    x = np.random.exponential(size=n)
    s = np.sum(x)
    return (s - n)/np.sqrt(n)

zs = [z(100) for _ in range(10000)]
plt.hist(zs, density=True, bins=50)
plt.show()

plt.hist(np.random.normal(size=10000), density=True, bins=50)
plt.show()
```

```python
def cauchy(y):
    return np.tan(np.pi * (y - 1/2))

x = [cauchy(np.random.random()) for _ in range(10000)]
plt.hist(x, range=(-10,10), bins=100, density=True)
plt.show()

y = np.array([ np.sum(x[:n]) / n for n in range(1,10000) ])
plt.plot(y)
plt.show()
```

```python
def cauchy_sum(n):
    x = np.random.standard_cauchy(size=n)
    return np.sum(x) / n

x = [cauchy_sum(10) for _ in range(10000)]
plt.hist(x, range=(-10,10), bins=50, density=True)
plt.show()

x = [cauchy_sum(100) for _ in range(10000)]
plt.hist(x, range=(-10,10), bins=50, density=True)
plt.show()
```
