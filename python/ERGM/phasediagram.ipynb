```python
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
```

```python
def I(p):
    if p <= 0 or p >= 1:
        return 0
    return 0.5 * p * np.log(p) + 0.5 * (1 - p) * np.log(1 - p)

def L(betas, edgecounts, p):
    return np.sum(betas * p**edgecounts) - I(p)
```

```python
pvals = np.linspace(0,1,100)

edgecounts = np.array([1,3,5])

beta1s = np.linspace(-5,0,10)
beta2s = np.linspace(0,5,100)
beta3s = np.linspace(0,5,100)

optimizers = np.zeros((len(beta1s),len(beta2s),len(beta3s)))

for i1, beta1 in tqdm(enumerate(beta1s)):
    for i2, beta2 in enumerate(beta2s):
        for i3, beta3 in enumerate(beta3s):
            betas = np.array([beta1,beta2,beta3])
            Lvals = np.array([L(betas,edgecounts,p) for p in pvals])
            optimizers[i1,i2,i3] = pvals[np.argmax(Lvals)]
```

```python
for i in range(len(beta1s)):
    plt.matshow(optimizers[i])
plt.colorbar()
```
