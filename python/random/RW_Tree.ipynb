```python
import numpy as np
import numpy.random as ra
```

```python
N = 500000
data = np.zeros(N)

for i in range(N):
    v = ""
    target = "00000"
    count = 0

    while len(v) < 20:
        r = ra.random()
        if r < 1/3:
            v = v[:-1]
        elif r < 2/3:
            v = v + "0"
        else:
            v = v + "1"

        if v == target:
            count += 1
        
    data[i] = count

print(np.average(data))
```
