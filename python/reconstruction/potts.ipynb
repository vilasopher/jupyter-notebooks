```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
beta = 2
q = 5

weights = np.array([
        [
            [ 
                np.exp(beta * ((1 if l == x else 0) + (1 if r == x else 0)))
                for l in range(q)
            ]
            for r in range(q)
        ]
        for x in range(q)
    ])

Z = np.sum(weights, axis=0)

probs = weights / Z

def combine_densities(fl, fr):
    f = np.transpose(fl) @ weights @ fr
    return f / np.sum(f)
```

```python
f = np.zeros(q)
f[0] = 1

for _ in range(2000):
    f = combine_densities(f,f)

f
```

```python
fig, axes = plt.subplots(1,q)
for i, ax in enumerate(axes):
    ax.matshow(weights[i])

fig, axes = plt.subplots(1,q)
for i, ax in enumerate(axes):
    ax.matshow(probs[i])
```
