```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(20,5))
N = 10000

for _ in range(1000):
    X = np.random.exponential(0.5,N)
    Y = np.random.exponential(0.5,N)
    Z = np.cumsum(X-Y)
    plt.plot(Z)

plt.show()
```

```python
dx = 0.01
T = 100

plt.figure(figsize=(20,5))

for _ in range(1000):
    B = np.cumsum(np.random.normal(scale=np.sqrt(dx), size=int(T/dx)))
    plt.plot(B)

plt.show()
```
